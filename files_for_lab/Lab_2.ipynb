```python
import pandas as pd
file1 = pd.read_csv("file1.csv")
file2 = pd.read_csv("file2.csv")
file3 = pd.read_csv("file3.csv")
```

```python
#Check headars are eaual, they are not
file1_columns = list(file1.columns)
file2_columns = list(file2.columns)
file3_columns = list(file3.columns)

print(file3_columns == file1_columns)
print(file2_columns == file1_columns)
print(file2_columns == file3_columns)
```

```python
#Total Claim Amount, st/State and  Gender is the culprit
print(file1_columns)
print(file2_columns)
print(file3_columns)
```

```python
#Fix and check


for i in range(len(file1_columns)):
    file1_columns[i].strip()
    file2_columns[i].strip()
    file3_columns[i].strip()
    
file1_columns.remove("GENDER")
file1_columns.insert(2,"Gender")
file2_columns.remove("GENDER")
file2_columns.insert(2,"Gender")

file1_columns.remove("ST")
file1_columns.insert(1,"State")
file2_columns.remove("ST")
file2_columns.insert(1,"State")

file1.columns = file1_columns 
file2.columns = file2_columns
file3.columns = file3_columns 

file1 = file1[file1_columns] 
file2 = file2[file1_columns]
file3 = file3[file1_columns]

file1_columns = list(file1.columns)
file2_columns = list(file2.columns)
file3_columns = list(file3.columns)

print(file1_columns)
print(file2_columns)
print(file3_columns)
print(file3_columns == file1_columns)
print(file2_columns == file1_columns)
print(file2_columns == file3_columns)

file2.head()



```

```python
df = pd.concat([file1,file2,file3])
```

```python
df.dtypes
```

```python
del df["Education"]
del df["Number of Open Complaints"]
df.head(10)
```

```python
#Convert CLV to float and remove NaNs
df["Customer Lifetime Value"] = df["Customer Lifetime Value"].apply (pd.to_numeric, errors='coerce')
df = df.dropna()
print(df.dtypes)

```

```python
df.head()
```

```python
CLV = df["Customer Lifetime Value"].tolist()
newCLV = []
for num in CLV:
    newCLV.append(num*100)
CLV = newCLV
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].replace(df["Customer Lifetime Value"].tolist(),CLV)

df.head()
```

```python
df = df[df["Income"] > 0]
```
